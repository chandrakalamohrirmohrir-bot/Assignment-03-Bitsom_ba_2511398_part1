```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]


for i in range(len(subjects)):
  subject = subjects[i]
  mark = marks[i]

  if mark>= 90:
    grade = "A+"
  elif mark>= 80:
    grade = "A"
  elif mark>= 70:
    grade = "B"
  elif mark>= 60:
    grade = "C"
  else:
    grade = "F"

  print(f"{subject} : {mark} ({grade})")

total = sum(marks)
print("Total marks:", total)

average = round(total/len(marks), 2)
print("Average:", average)

max_mark = max(marks)
index = marks.index(max_mark)
print("Highest:", subjects[index], "-", max_mark)

min_mark = min(marks)
index = marks.index(min_mark)
print("Lowest:", subjects[index], "-", min_mark)

new_subjects = []
new_marks = []

while True:
  subject = input("Enter subject (or 'done' to stop): ")

  if subject.lower() == "done":
    break

  mark_input = input("Enter mark (0-100):")

  if not mark_input.isdigit():
    print("Invalid marks! Must be a number.")
    continue

  mark = int(mark_input)

  if mark < 0 or mark > 100:
    print("Marks must be between 0 and 100.")
    continue

  new_subjects.append(subject)
  new_marks.append(mark)

print("new subjects added:",len(new_subjects))

all_marks = marks + new_marks
new_average = round(sum(all_marks)/ len(all_marks), 2)

print("Updated average:", new_average)



```

```python
class_data = [
    ("Ayesha Sharma", [88, 72, 95, 60, 78]),
    ("Rohit Verma", [55, 68, 49, 72, 61]),
    ("Priya Nair", [91, 85, 88, 94, 79]),
    ("Karan Mehta", [40, 55, 38, 62, 50]),
    ("Sneha Pillai", [75, 80, 70, 68, 85]),
]

print("Name            | Average | Status")
print("--------------------------------------")

for name, marks in class_data:
  avg = round(sum(marks) / len(marks), 2)

  if avg >= 60:
    status = "Pass"
  else:
    status = "Fail"

  print(f"{name:<15} | {avg:<7} | {status}")

pass_count = 0
fail_count = 0

for name, marks in class_data:
  avg = round(sum(marks) / len(marks), 2)

  if avg >= 60:
    pass_count += 1
  else:
    fail_count += 1

print("passed:", pass_count)
print("Failed:", fail_count)


top_name = ""
top_avg = 0

for name, marks in class_data:
  avg = round(sum(marks) / len(marks), 2)

  if avg > top_avg:
    top_avg = avg
    top_name = name

print("Topper:", top_name,"-", top_avg)

total_avg = 0

for name, marks in class_data:
  avg = round(sum(marks) / len(marks), 2)
  total_avg += avg

class_avg = round(total_avg / len(class_data), 2)

print("Class Average:", class_avg)

```
